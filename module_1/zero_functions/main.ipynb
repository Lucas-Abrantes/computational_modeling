```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x, y, label='y = x^2')
plt.plot(x, y,'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('y = x^2')
plt.grid()
plt.show()
```

```python
def func_1(x):
    return np.exp(-x**2)*np.cos(2*np.pi*x)

def func_2(x):
    return np.cos(2*np.pi*x)

x1 = np.linspace(0, 5, 100)

x2 = np.linspace(0, 5, 100)

y1 = func_1(x1)
y2 = func_2(x2)

plt.figure()
plt.subplot(2, 1, 1)
plt.title('exp(-x^2)*cos(2*pi*x)')
plt.plot(x1, y1, 'k')
plt.plot(x1, y1, 'ro')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g--')
plt.title('cos(2*pi*x)')
plt.grid()

plt.tight_layout()  # Ajusta o espaçamento entre os subplots
plt.show()
```
