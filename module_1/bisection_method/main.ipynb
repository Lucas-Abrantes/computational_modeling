```python
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    # return np.log10(x) * x - 1
    return x**3 - 2*x**2 + x - 1
x = np.arange(1, 3, 0.1)
y = func(x)

b = 2.0 # intervalo superior
a = 1.0 # intervalo inferior
esp = 0.0001 # precisão
x0 = 0

while (b-a) > esp:
    x0 = (a + b) / 2
    if func(x0) * func(b) > 0:
        b = x0
    else:
        a = x0


# pode ser usado para definir o estilo do gráfico
# pode colocar tudo na mesma função plot ou separado
# plt.plot(x, y, 'r--')
# plt.plot(x, y, 'bo') 
plt.figure()
plt.plot(x, y, 'b--')


plt.plot(x0, func(x0), 'ro')
plt.text(x0, 0.5*func(x0), '$x_0$', fontsize=12)
plt.grid()
print('[', a, ',', b, ']')
plt.show()
```
