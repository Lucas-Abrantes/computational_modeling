```python

import numpy as np

def f(x):
    return (1-np.exp(-2*x))

a = float(input("Digite o intervalo inferior: "))

b= float(input("Digite o intervalo superior: "))

n = int(input("Digite o número de partições: "))

deltax = (b-a)/n

#Este código é o responsável pelo cálculo da regra do trapézio repetida
soma1=0
soma2=0
for i in range(0,n + 1,1):
    x=a+deltax*i
    if (i==0 or i==n+1):
        soma1 = (soma1 + f(x))
    else:
        soma2 = (soma2 + 2*f(x))
    soma = (soma1 + soma2)*(deltax/2)
print(soma)
```

#### Melhorar a regra do trapézio quando não se tem a função definida, mas apenas os pontos. Ou seja, eu vou ter uma tabela com os valores de x e f(x)

```python
import matplotlib.pyplot as plt

def trapezoidal_from_points(x, y):
    if len(x) != len(y):
        raise ValueError("x e y devem ter o mesmo comprimento.")
    
    total = 0.0
    for i in range(len(x) - 1):
        h = x[i+1] - x[i]
        total += (y[i] + y[i+1]) * h / 2
    return total

# Dados
x = [0., 0.13793103, 0.27586207, 0.4137931, 0.55172414, 0.68965517,
     0.82758621, 0.96551724, 1.10344828, 1.24137931, 1.37931034, 1.51724138,
     1.65517241, 1.79310345, 1.93103448, 2.06896552, 2.20689655, 2.34482759,
     2.48275862, 2.62068966, 2.75862069, 2.89655172, 3.03448276, 3.17241379,
     3.31034483, 3.44827586, 3.5862069, 3.72413793, 3.86206897, 4.]
y = [0.        , 0.2410824 , 0.42404407, 0.56289691, 0.66827477,
     0.74824788, 0.80894089, 0.85500188, 0.88995837, 0.91648747,
     0.93662087, 0.95190046, 0.96349642, 0.97229679, 0.97897554,
     0.98404417, 0.98789084, 0.99081015, 0.99302566, 0.99470705,
     0.99598309, 0.99695149, 0.99768643, 0.99824419, 0.99866749,
     0.99898873, 0.99923253, 0.99941756, 0.99955797, 0.99966454]  

resultado = trapezoidal_from_points(x, y)
print("Integral aproximada com tabela:", resultado)

# Gráfico da área aproximada
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Curva f(x)')
plt.fill_between(x, y, color='skyblue', alpha=0.4, label='Área (Regra do Trapézio)')
for i in range(len(x) - 1):
    plt.plot([x[i], x[i], x[i+1], x[i+1]], [0, y[i], y[i+1], 0], 'r-', alpha=0.5)
plt.title("Visualização da Regra do Trapézio com Pontos Tabelados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
