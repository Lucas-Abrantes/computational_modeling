```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
g0 = 9.81         # aceleração gravitacional na superfície (m/s²)
R = 6.37e6        # raio da Terra (m)
v0 = 1400         # velocidade inicial (m/s)
x0 = 0            # posição inicial (m)

# Parâmetros de simulação
t_final = 1000    # tempo máximo (s)
h = 0.1           # passo de tempo (s)
n = int(t_final / h)

# Inicialização dos vetores
t = np.zeros(n)
x = np.zeros(n)
v = np.zeros(n)

# Condições iniciais
x[0] = x0
v[0] = v0

# Método de Euler para sistemas
for i in range(1, n):
    a = -g0 * (R**2) / (R + x[i-1])**2  # aceleração no ponto atual
    v[i] = v[i-1] + h * a
    x[i] = x[i-1] + h * v[i-1]
    t[i] = t[i-1] + h
    
    if v[i] <= 0:  # o ponto de altura máxima é onde a velocidade se anula
        print(f"Altura máxima atingida: {x[i]:.2f} m no tempo {t[i]:.2f} s")
        break

# Gráficos
plt.plot(t[:i+1], x[:i+1])
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.title('Altura vs Tempo (Método de Euler)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros físicos e da malha
N = 100  # Tamanho do grid (N x N)
h = 1.0  # Espaçamento da malha
q = 1.0  # Valor da carga (em unidades arbitrárias)
epsilon0 = 1.0  # Permissividade (também arbitrária)
erro = 1e-5  # Critério de convergência

# Inicializa o potencial e a densidade de carga
phi = np.zeros((N+1, N+1))
rho = np.zeros((N+1, N+1))

# Posições das duas cargas (carga positiva e carga negativa)
i1, j1 = 25, 50  # Carga +q
i2, j2 = 75, 50  # Carga -q

# Define a densidade de carga na malha (discretização da função delta)
rho[i1, j1] = q / h**2
rho[i2, j2] = -q / h**2

# Método iterativo de Gauss-Seidel
delta = 1.0
iteracoes = 0

while delta > erro:
    delta = 0.0
    for i in range(1, N):
        for j in range(1, N):
            phi_old = phi[i, j]
            phi[i, j] = 0.25 * (phi[i+1, j] + phi[i-1, j] + phi[i, j+1] + phi[i, j-1] + h**2 * rho[i, j] / epsilon0)
            diff = abs(phi[i, j] - phi_old)
            if diff > delta:
                delta = diff
    iteracoes += 1

print(f"Convergiu em {iteracoes} iterações.")

# Visualização do potencial elétrico
plt.figure(figsize=(4, 4))
plt.contourf(phi, levels=50, cmap='coolwarm')
plt.colorbar(label='Potencial φ')
plt.title("Potencial elétrico gerado por duas cargas puntiformes")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema de EDOs:
# y1' = y2
# y2' = x^2

def f(x, y1, y2):
    dy1dx = y2
    dy2dx = x**2
    return dy1dx, dy2dx

# Condições iniciais
x0 = 0
y1_0 = 2 # y(0)
y2_0 = 0 # y'(0), assumido como 0

x_final = 5
h = 0.1
n = int((x_final - x0)/h) + 1

x = np.linspace(x0, x_final, n)
y1 = np.zeros(n)
y2 = np.zeros(n)

y1[0] = y1_0
y2[0] = y2_0

# Método de Runge-Kutta de 4ª ordem
for i in range(1, n):
    k1_y1, k1_y2 = f(x[i-1], y1[i-1], y2[i-1])
    k2_y1, k2_y2 = f(x[i-1] + h/2, y1[i-1] + h/2 * k1_y1, y2[i-1] + h/2 * k1_y2)
    k3_y1, k3_y2 = f(x[i-1] + h/2, y1[i-1] + h/2 * k2_y1, y2[i-1] + h/2 * k2_y2)
    k4_y1, k4_y2 = f(x[i-1] + h, y1[i-1] + h * k3_y1, y2[i-1] + h * k3_y2)

    y1[i] = y1[i-1] + (h/6)*(k1_y1 + 2*k2_y1 + 2*k3_y1 + k4_y1)
    y2[i] = y2[i-1] + (h/6)*(k1_y2 + 2*k2_y2 + 2*k3_y2 + k4_y2)

# Resultado
print(f"y({x[5]:.2f}) = {y1[5]}")

# Gráfico
plt.plot(x, y1, label="y(x) - Solução aproximada")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solução de y'' = x² com y(0)=2, y'(0)=0 (RK4)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do modelo
N = 763  # População total
I0, R0 = 1, 0  # Infectados e recuperados iniciais
S0 = N - I0 - R0  # Suscetíveis iniciais
beta, gamma = 1.68, 0.44  # Taxa de contato e taxa de recuperação

# Intervalo de tempo
t = np.linspace(0, 14, 160)
dt = t[1] - t[0]  # passo de tempo
n = len(t)

# Inicialização dos vetores
S = np.zeros(n)
I = np.zeros(n)
R = np.zeros(n)

# Condições iniciais
S[0] = S0
I[0] = I0
R[0] = R0

# Função derivadas do modelo SIR
def deriv(s, i, r, beta, gamma, N):
    dSdt = -beta * s * i / N
    dIdt = beta * s * i / N - gamma * i
    dRdt = gamma * i
    return dSdt, dIdt, dRdt

# Método de Runge-Kutta 4ª ordem
for k in range(n - 1):
    k1_S, k1_I, k1_R = deriv(S[k], I[k], R[k], beta, gamma, N)
    k2_S, k2_I, k2_R = deriv(S[k] + dt*k1_S/2, I[k] + dt*k1_I/2, R[k] + dt*k1_R/2, beta, gamma, N)
    k3_S, k3_I, k3_R = deriv(S[k] + dt*k2_S/2, I[k] + dt*k2_I/2, R[k] + dt*k2_R/2, beta, gamma, N)
    k4_S, k4_I, k4_R = deriv(S[k] + dt*k3_S, I[k] + dt*k3_I, R[k] + dt*k3_R, beta, gamma, N)

    S[k+1] = S[k] + (dt/6)*(k1_S + 2*k2_S + 2*k3_S + k4_S)
    I[k+1] = I[k] + (dt/6)*(k1_I + 2*k2_I + 2*k3_I + k4_I)
    R[k+1] = R[k] + (dt/6)*(k1_R + 2*k2_R + 2*k3_R + k4_R)

# Plotagem dos resultados
plt.figure(figsize=(6, 6))
plt.plot(t, S/1000, 'b', label='Suscetíveis')
plt.plot(t, I/1000, 'r', label='Infectados')
plt.plot(t, R/1000, 'g', label='Recuperados')
plt.title('Modelo SIR com Runge-Kutta de 4ª Ordem')
plt.xlabel('Tempo (dias)')
plt.ylabel('Proporção da População')
plt.ylim(0, 1.2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1. / 10
# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
plt.title('Modelo SIR')
plt.figure(figsize=(8, 8))
ax.plot(t, S / 1000, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I / 1000, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R / 1000, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 1) Equação de Laplace em 2D via "pseudo-tempo" MoL
#    u_t = u_xx + u_yy, com u fixo nas bordas,
#    e integramos até a solução estacionária.
# ----------------------------------------------------

def solve_laplace_via_mol(Nx=50, Ny=50, t_final=5.0):
    # Domínio unitário [0,1]×[0,1]
    dx, dy = 1/(Nx-1), 1/(Ny-1)
    x = np.linspace(0,1,Nx)
    y = np.linspace(0,1,Ny)
    X, Y = np.meshgrid(x,y,indexing='ij')
    
    # Condições de contorno: u(x,0)=sin(pi x), outras = 0
    u0 = np.zeros((Nx,Ny))
    u0[:,0] = np.sin(np.pi * x)
    
    # Monta o operador Laplaciano 2D em formato esparso
    Ix = sp.eye(Nx); Iy = sp.eye(Ny)
    e = np.ones(Nx)
    D2x = sp.diags([e,-2*e,e],[-1,0,1],shape=(Nx,Nx)) / dx**2
    e = np.ones(Ny)
    D2y = sp.diags([e,-2*e,e],[-1,0,1],shape=(Ny,Ny)) / dy**2
    
    L2D = sp.kronsum(D2x, D2y, format='csr')
    
    # Função RHS para u' = L u, mas impondo BCs em cada passo
    def rhs_lap(t, u_flat):
        u = u_flat.reshape((Nx,Ny))
        # enforça BCs
        u[0,:] = 0; u[-1,:] = 0; u[:,0] = np.sin(np.pi*x); u[:,-1] = 0
        du = L2D.dot(u.flatten())
        return du
    
    sol = solve_ivp(rhs_lap, [0, t_final], u0.flatten(),
                    method='RK23', atol=1e-6, rtol=1e-6)
    U = sol.y[:,-1].reshape((Nx,Ny))
    
    # Plot
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, U, cmap='viridis')
    ax.set_title('Solução estacionária de Laplace via MoL')
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u')
    plt.show()

# ----------------------------------------------------
# 2) Equação do Calor em 1D: u_t = α u_xx
#    Domínio x∈[0,1], BCs u(0)=u(1)=0, 
#    CI: u(x,0)=sin(pi x). α=0.1
# ----------------------------------------------------

def solve_heat_1d_mol(N=50, α=0.1, t_span=(0,1.0)):
    dx = 1/(N-1)
    x = np.linspace(0,1,N)
    # Monta matriz de segunda derivada 1D
    e = np.ones(N)
    D2 = sp.diags([e,-2*e,e],[-1,0,1],shape=(N,N)) / dx**2
    # Impõe Dirichlet homogeneous: elimina linhas/cols de borda
    D2 = D2.tolil()
    D2[0,:] = 0; D2[-1,:] = 0
    
    def rhs_heat(t, u):
        # u[0]=u[-1]=0 fixos
        dudx2 = D2.dot(u)
        dudt = α * dudx2
        # mantém BCs
        dudt[0] = 0; dudt[-1] = 0
        return dudt
    
    # CI
    u0 = np.sin(np.pi * x)
    u0[0]=0; u0[-1]=0
    
    sol = solve_ivp(rhs_heat, t_span, u0, t_eval=[t_span[1]], 
                    method='DOP853', atol=1e-8, rtol=1e-6)
    u_final = sol.y[:,-1]
    
    # Plot
    plt.figure(figsize=(6,4))
    plt.plot(x, u_final, '-o')
    plt.title(f'Calor 1D via MoL em t={t_span[1]:.2f}')
    plt.xlabel('x'); plt.ylabel('u(x,t)')
    plt.grid(True)
    plt.show()

# ----------------------------------------------------
# Executa exemplos
# ----------------------------------------------------
if __name__ == "__main__":
    solve_laplace_via_mol()
    solve_heat_1d_mol()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Função que define o sistema de EDOs
def laplace_odes(tau, T_internal_flat, N, h, T_bc_values):
    """
    Calcula a RHS (dT/dtau) para o sistema de EDOs da equação de Laplace.
    tau: pseudo-tempo (não usado explicitamente na equação de Laplace estacionária, mas requerido por solve_ivp)
    T_internal_flat: array 1D contendo os valores de temperatura nos pontos internos da malha.
    N: número de pontos internos na direção x (e y, para uma malha quadrada).
    h: espaçamento da malha.
    T_bc_values: dicionário com os valores de temperatura nas fronteiras.
                 {'bottom': val, 'top': val, 'left': val, 'right': val}
    """
    T_internal = T_internal_flat.reshape((N, N))
    
    # Inicializa a malha completa (incluindo fronteiras)
    T_full = np.zeros((N + 2, N + 2))
    
    # Aplica condições de contorno
    T_full[1:N+1, 0] = T_bc_values['bottom']  # Base
    T_full[1:N+1, -1] = T_bc_values['top']   # Topo
    T_full[0, 1:N+1] = T_bc_values['left']   # Esquerda
    T_full[-1, 1:N+1] = T_bc_values['right']  # Direita

    # Cantos (importante para consistência se BCs variam)
    T_full[0, 0] = T_bc_values['left'] 
    T_full[-1, 0] = T_bc_values['right']
    T_full[0, -1] = T_bc_values['left'] 
    T_full[-1, -1] = T_bc_values['right']

    # Preenche a parte interna da malha com os valores atuais de T
    T_full[1:N+1, 1:N+1] = T_internal
    
    dTdtau_internal = np.zeros((N, N))
    
    # Calcula dT/dtau para os pontos internos
    for i in range(N):
        for j in range(N):
            i_full = i + 1
            j_full = j + 1
            
            T_center = T_full[i_full, j_full]
            T_north = T_full[i_full, j_full + 1]
            T_south = T_full[i_full, j_full - 1]
            T_east = T_full[i_full + 1, j_full]
            T_west = T_full[i_full - 1, j_full]
            
            laplacian = (T_east + T_west + T_north + T_south - 4 * T_center) / (h**2)
            dTdtau_internal[i, j] = laplacian
            
    return dTdtau_internal.flatten()

# Evento para parar a integração quando o estado estacionário é alcançado
def steady_state_event(tau, T_internal_flat, N_func_arg, h_func_arg, T_bc_values_func_arg, tolerance_func_arg):
    # N_func_arg, h_func_arg, T_bc_values_func_arg são passados por solve_ivp via args
    # tolerance_func_arg é capturado pela lambda do escopo externo
    dTdtau_flat = laplace_odes(tau, T_internal_flat, N_func_arg, h_func_arg, T_bc_values_func_arg)
    max_rate_of_change = np.max(np.abs(dTdtau_flat))
    return max_rate_of_change - tolerance_func_arg
steady_state_event.terminal = True 
steady_state_event.direction = -1  

# Parâmetros da simulação
N = 20  
h = 1.0 / (N + 1)  

T_bc_values = {
    'bottom': 0.0,
    'top': 100.0,
    'left': 0.0,
    'right': 0.0
}

T0_internal = np.zeros((N, N))
T0_internal_flat = T0_internal.flatten()

tau_start = 0.0
tau_end = 2.0  
integration_tolerance = 1e-3 

# Argumentos para laplace_odes (e para a parte de steady_state_event que é comum)
common_args = (N, h, T_bc_values)

print(f"Iniciando a simulação com N={N}, h={h:.4f}")
print(f"Condições de contorno: Topo={T_bc_values['top']}, Base={T_bc_values['bottom']}, Esq={T_bc_values['left']}, Dir={T_bc_values['right']}")

# Resolve o sistema de EDOs
# A lambda para 'events' agora aceita os argumentos de 'args' e passa 'integration_tolerance' do escopo.
solution = solve_ivp(
    fun=laplace_odes,
    t_span=(tau_start, tau_end),
    y0=T0_internal_flat,
    method='RK45',  
    args=common_args,
    events=lambda t, y, N_lambda_arg, h_lambda_arg, T_bc_lambda_arg: steady_state_event(t, y, N_lambda_arg, h_lambda_arg, T_bc_lambda_arg, integration_tolerance),
    dense_output=True 
)

print(f"Simulação concluída em pseudo-tempo tau = {solution.t[-1]:.4f}")
if solution.status == 1: 
    print("Estado estacionário alcançado.")
elif solution.status == 0: 
    print("Atenção: O fim do intervalo de integração foi alcançado antes do estado estacionário (de acordo com a tolerância).")
else:
    print(f"Solver terminou com status: {solution.status}")

T_final_internal_flat = solution.y[:, -1]
T_final_internal = T_final_internal_flat.reshape((N, N))

T_final_full = np.zeros((N + 2, N + 2))
T_final_full[1:N+1, 0] = T_bc_values['bottom']
T_final_full[1:N+1, -1] = T_bc_values['top']
T_final_full[0, 1:N+1] = T_bc_values['left']
T_final_full[-1, 1:N+1] = T_bc_values['right']
T_final_full[0, 0] = T_bc_values['left'] 
T_final_full[-1, 0] = T_bc_values['right']
T_final_full[0, -1] = T_bc_values['left'] 
T_final_full[-1, -1] = T_bc_values['right']
T_final_full[1:N+1, 1:N+1] = T_final_internal

x_coords = np.linspace(0, 1, N + 2)
y_coords = np.linspace(0, 1, N + 2)
X, Y = np.meshgrid(x_coords, y_coords)

plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, T_final_full, levels=50, cmap='jet')
plt.colorbar(contour, label='Temperatura (°C)')
plt.title('Distribuição de Temperatura (Equação de Laplace - Método das Linhas)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()

final_dTdtau_flat = laplace_odes(solution.t[-1], T_final_internal_flat, N, h, T_bc_values)
max_final_rate = np.max(np.abs(final_dTdtau_flat))
print(f"Taxa máxima de variação final (max|dT/dtau|): {max_final_rate:.2e}")

if solution.status ==1: # Somente se o evento parou a simulação
    if max_final_rate <= integration_tolerance:
        print("Confirmação: Critério de estado estacionário satisfeito.")
    else:
        # Isso pode acontecer se a tolerância do root finder do evento for diferente da nossa verificação
        print(f"Atenção: A taxa final ({max_final_rate:.2e}) é ligeiramente diferente da tolerância ({integration_tolerance:.2e}), mas o evento terminal ocorreu.")
elif solution.status == 0:
     print("Critério de estado estacionário não satisfeito pois o tempo máximo de integração foi atingido.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da malha
N = 100  # Tamanho da malha (N x N)
max_iter = 5000  # Número máximo de iterações
tolerancia = 1e-5  # Critério de convergência

# Inicialização da matriz de potencial
phi = np.zeros((N, N))

phi[0, :] = 100  # Borda superior em 100 °C
# Demais bordas já estão em 0 por inicialização

# Cópia para armazenar o estado anterior
phi_old = np.zeros((N, N))

# Loop de iteração
for iter in range(max_iter):
    phi_old[:, :] = phi  # Copiar os valores atuais para a matriz antiga
    # Atualizar os pontos internos
    for i in range(1, N-1):
        for j in range(1, N-1):
            phi[i, j] = 0.25 * (phi_old[i+1, j] + phi_old[i-1, j] + phi_old[i, j+1] + phi_old[i, j-1])
    # Reaplicar condições de contorno (garantir bordas fixas)
    phi[0, :] = 100  # Borda superior
    phi[-1, :] = 0   # Borda inferior
    phi[:, 0] = 0    # Borda esquerda
    phi[:, -1] = 0   # Borda direita
    
    # Verificar convergência
    diff = np.max(np.abs(phi - phi_old))
    if diff < tolerancia:
        print(f"Convergência alcançada na iteração {iter+1}")
        break

plt.figure(figsize=(8, 6))
plt.imshow(phi, origin='lower', cmap='jet', interpolation='bicubic', extent=[0, N, 0, N])
plt.colorbar(label='Temperatura (°C)')
plt.title('Solução da Equação de Laplace usando o Método das Linhas')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do problema
nx, ny = 100, 100        # número de pontos na malha
max_iter = 5000       # número máximo de iterações
tolerance = 1e-6       # tolerância para convergência

# Inicializa o domínio com zeros
phi = np.zeros((ny, nx))

# Condições de contorno
phi[:, 0] = 100  # Borda esquerda
phi[:, -1] = 0   # Borda direita
phi[0, :] = 0    # Borda inferior
phi[-1, :] = 0   # Borda superior

# Método iterativo (relaxação de Jacobi)
def solve_laplace(phi, max_iter, tolerance):
    for it in range(max_iter):
        phi_old = phi.copy()
        for i in range(1, ny - 1):
            for j in range(1, nx - 1):
                phi[i, j] = 0.25 * (phi_old[i+1, j] + phi_old[i-1, j] + phi_old[i, j+1] + phi_old[i, j-1])
        # Checagem de convergência
        error = np.max(np.abs(phi - phi_old))
        if error < tolerance:
            print(f"Convergência atingida após {it} iterações com erro {error:.2e}")
            break
    return phi

# Resolver
phi = solve_laplace(phi, max_iter, tolerance)
phi_rot = phi.T
# Visualização
plt.figure(figsize=(8, 6))
plt.imshow(phi_rot, origin='lower', cmap='jet', interpolation='bicubic', extent=[0, nx, 0, ny])
plt.colorbar(label='Temperatura (°C)')
plt.title('Solução da Equação de Laplace (Método das Linhas)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do modelo
N = 990  # População total
I0, R0 = 10, 0  # Infectados e recuperados iniciais
S0 = N - I0 - R0  # Suscetíveis iniciais
beta, gamma = 0.2, 0.1  # Taxa de contato e taxa de recuperação

# Intervalo de tempo
t = np.linspace(0, 300, 160)
dt = t[1] - t[0]  # Passo de tempo
n = len(t)

# Inicialização dos vetores
S = np.zeros(n)
I = np.zeros(n)
R = np.zeros(n)

# Condições iniciais
S[0] = S0
I[0] = I0
R[0] = R0

# Função das derivadas do modelo SIR
def deriv(s, i, r, beta, gamma, N):
    dSdt = -beta * s * i / N
    dIdt = beta * s * i / N - gamma * i
    dRdt = gamma * i
    return dSdt, dIdt, dRdt

# Método de Runge-Kutta 4ª ordem
for k in range(n - 1):
    k1_S, k1_I, k1_R = deriv(S[k], I[k], R[k], beta, gamma, N)
    k2_S, k2_I, k2_R = deriv(
        S[k] + dt * k1_S / 2, I[k] + dt * k1_I / 2, R[k] + dt * k1_R / 2, beta, gamma, N
    )
    k3_S, k3_I, k3_R = deriv(
        S[k] + dt * k2_S / 2, I[k] + dt * k2_I / 2, R[k] + dt * k2_R / 2, beta, gamma, N
    )
    k4_S, k4_I, k4_R = deriv(
        S[k] + dt * k3_S, I[k] + dt * k3_I, R[k] + dt * k3_R, beta, gamma, N
    )

    S[k + 1] = S[k] + (dt / 6) * (k1_S + 2 * k2_S + 2 * k3_S + k4_S)
    I[k + 1] = I[k] + (dt / 6) * (k1_I + 2 * k2_I + 2 * k3_I + k4_I)
    R[k + 1] = R[k] + (dt / 6) * (k1_R + 2 * k2_R + 2 * k3_R + k4_R)

# Plotagem dos resultados
plt.figure(figsize=(8, 4))
plt.plot(t, S / N, 'b', label='Suscetíveis')
plt.plot(t, I / N, 'r', label='Infectados')
plt.plot(t, R / N, 'g', label='Recuperados')
plt.title('Modelo SIR com Runge-Kutta de 4ª Ordem')
plt.xlabel('Tempo (dias)')
plt.ylabel('Proporção da População')
plt.ylim(0, 1.2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

 

# Parâmetros do modelo
N = 990  # População total
I0, R0 = 10, 0  # Infectados e recuperados iniciais
S0 = N - I0 - R0  # Suscetíveis iniciais
beta, gamma = 0.0005, 0.1  # Taxa de contato e taxa de recuperação
# A grid of time points (in days)
t = np.linspace(0, 160,160)

 

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

 

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(figsize=(8, 6), facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
plt.title('Modelo SIR')
ax.plot(t, S / 1000, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I / 1000, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R / 1000, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time / days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(visible=True, which='major', color='w', linewidth=2, linestyle='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
